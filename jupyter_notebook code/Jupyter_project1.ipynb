```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Define a function to calculate VIF
def calc_vif(df):
    # Step 1: Create an empty DataFrame to store VIF results
    vif = pd.DataFrame()

    # Step 2: Assign column names to a 'variables' column in VIF DataFrame
    vif["variables"] = df.columns

    # Step 3: Calculate VIF for each column using a list comprehension
    vif["VIF"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]

    # Step 4: Return the resulting VIF DataFrame
    return vif

# Call the function to calculate VIF for 'new_df'
calc_vif(new_df)
```

```python
#standardscaler standardizes that normalised the data set by scaling features to have :
# mean = 0 and Standard deviation =1
from sklearn.preprocessing import StandardScaler
#Splits the dataset into training and testing subsets.
from sklearn.model_selection import train_test_split
#RandomForestClassifier is an ensemble method that uses multiple decision trees to improve accuracy and reduce overfitting.
#Combines predictions from many decision trees.
#Uses bagging: Samples random subsets of data for each tree.
from sklearn.ensemble import RandomForestClassifier
#Builds a single decision tree for classification problems.
from sklearn.tree import DecisionTreeClassifier

import itertools
from collections import Counter
import sklearn.metrics as metrics
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
from sklearn.preprocessing import StandardScaler

# Check if 'Actual_amount_orig' exists and apply scaling
if 'Actual_amount_orig' in new_df.columns:
    scaler = StandardScaler()
    new_df["NormalizedAmount"] = scaler.fit_transform(new_df["Actual_amount_orig"].values.reshape(-1, 1))
    new_df.drop(["Actual_amount_orig"], inplace=True, axis=1)
else:
    print("Column 'Actual_amount_orig' not found in DataFrame!")

# Verify DataFrame
print(new_df.head())
print(new_df.columns)
```

```python
# Perform Scaling
scaler = StandardScaler()
new_df["NormalizedAmount"] = scaler.fit_transform(new_df["amount"].values.reshape(-1, 1))
new_df.drop(["amount"], inplace= True, axis= 1)

Y = new_df["isFraud"]
X = new_df.drop(["isFraud"], axis= 1)


# Split the data
(X_train, X_test, Y_train, Y_test) = train_test_split(X, Y, test_size= 0.3, random_state= 42)

print("Shape of X_train: ", X_train.shape)
print("Shape of X_test: ", X_test.shape)
```

```python
# 1. Create an instance of the DecisionTreeClassifier
decision_tree = DecisionTreeClassifier()
#lets train the decision tree model using the training data 
decision_tree.fit(X_train , Y_train)
#Predict the target labels for the test data using the trained model
Y_pred_dt = decision_tree.predict(X_test)
#calculate the acc of decision tree model on the test data
decision_tree_score = decision_tree.score(X_test, Y_test)* 100
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create an instance of the RandomForestClassifier with 50 trees and parallel processing
random_forest = RandomForestClassifier(n_estimators=50, n_jobs=-1, max_depth=10)

# Train the Random Forest model using the training data
random_forest.fit(X_train, Y_train)

# Predict the target labels for the test data using the trained model
Y_pred_rf = random_forest.predict(X_test)

# Calculate the accuracy of the random forest model on the test data
random_forest_score = random_forest.score(X_test, Y_test) * 100

print(f"Random Forest Accuracy: {random_forest_score:.2f}%")
```

```python
# Print scores of our classifiers

print("Decision Tree Score: ", decision_tree_score)
print("Random Forest Score: ", random_forest_score)
```

```python
from sklearn.metrics import confusion_matrix

# Calculate confusion matrix
tn, fp, fn, tp = confusion_matrix(Y_test, Y_pred_dt).ravel()

# Print confusion matrix components
print("TP, FP, TN, FN - Decision Tree")

print(f'True Positives: {tp}') 
print(f'False Positives: {fp}')
print(f'True Negatives: {tn}')
print(f'False Negatives: {fn}')

# Optional: Print the confusion matrix as a whole for better context
print("\nConfusion Matrix:")
print(confusion_matrix(Y_test, Y_pred_dt))

print("----------------------------------------------------------------------------------------")
```

```python
# key terms of Confusion Matrix - RF

print("TP,FP,TN,FN - Random Forest")
tn, fp, fn, tp = confusion_matrix(Y_test, Y_pred_rf).ravel()
print(f'True Positives: {tp}')
print(f'False Positives: {fp}')
print(f'True Negatives: {tn}')
print(f'False Negatives: {fn}')
```
